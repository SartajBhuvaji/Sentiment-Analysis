```python
import os
import sys
import json
import emoji
import requests
import pandas as pd
from dotenv import load_dotenv
load_dotenv()
```

```python
youtube_api_key = os.getenv("YOUTUBE_API_KEY")
max_results = os.getenv("MAX_RESULTS")
video_id = 'kffacxfA7G4'
```

```python
url = f'https://www.googleapis.com/youtube/v3/commentThreads?key={youtube_api_key}&textFormat=plainText&part=snippet&videoId={video_id}&maxResults={max_results}'
```

```python
# use the URL to get the JSON data
response = requests.get(url)
response.encoding = 'utf-8'
data = response.json()
```

```python
text_display = []
like_count = []

for comment in data["items"]:
    snippet = comment["snippet"]["topLevelComment"]["snippet"]
    text = snippet["textDisplay"]
    
    # Replace emojis with their corresponding text representations
    #text = emoji.demojize(text)
    
    text_display.append(text)
    like_count.append(snippet["likeCount"])

# Create a DataFrame
df = pd.DataFrame({'comment': text_display, 'likeCount': like_count})
```

```python
def is_emoji(char):
    return char in emoji.UNICODE_EMOJI
```

```python
df['comment']
```

```python
emojis = []
emoji_pattern = re.compile(emoji.get_emoji_regexp())

# Create a new column to store comments without emojis
df['comment_no_emojis'] = df['comment'].apply(lambda x: emoji_pattern.sub('', x))

for comment in df['comment']:
    emojis_found = emoji_pattern.findall(comment)
    emojis.append(emojis_found)
```

```python
df['emojis'] = emojis  
df.drop(columns=['comment'], inplace=True)
```

```python
df.head(5)
```

```python
df['comment_no_emojis'] = df['comment_no_emojis'].str.replace(r'[^a-zA-Z]+', ' ')
```

```python
df.sort_values(by=['likeCount'], ascending=False, inplace=True) 
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm

sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores(df['comment'][2])
```

```python
#Loop through all comments and append the scores to a list
scores = []

for comment in df['comment_no_emojis']:
    score = sia.polarity_scores(comment)
    scores.append(score)

# Convert the list to a DataFrame
scores_df = pd.DataFrame(scores)

# Join the scores DataFrame to the original one
df = df.join(scores_df)
```
